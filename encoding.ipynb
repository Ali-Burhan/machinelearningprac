```python
import pandas as pd
dataset = pd.read_csv('./dataset.csv')
```

```python
sex = dataset['Sex']
```

```python
sex
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Male','female']])
```

```python
result = oe.fit_transform(sex.values.reshape(-1,1))
```

```python
#Using Sklearn Library
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
#working with onehotencoder you get a sparse array
#you make toarray method to convert it into Array
#then again concat it into Dataframe
result = ohe.fit_transform(dataset[['Sex']]).toarray()
```
