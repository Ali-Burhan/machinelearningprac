```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('../students.csv')
```

```python
import seaborn as sns
sns.kdeplot(dataset,x='CGPA')
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
CGPA_COL = dataset[['CGPA']]
```

```python
transformer_col = FunctionTransformer(func=np.log1p)
```

```python
transformed_data = transformer_col.fit_transform(CGPA_COL)
```

```python
sns.kdeplot(transformed_data,x='CGPA')
```

```python
transformer_col_2 = FunctionTransformer(lambda x: x**2)
```

```python
transformed_data_2 = transformer_col_2.fit_transform(CGPA_COL)
```

```python
sns.kdeplot(transformed_data_2,x='CGPA')
```

```python
transformer_col_2 = FunctionTransformer(lambda x: x**1/2)
transformed_data_2 = transformer_col_2.fit_transform(CGPA_COL)
sns.kdeplot(transformed_data_2,x='CGPA')
```

```python
transformer_col_3 = FunctionTransformer(lambda x: 1/(x+0.1))
transformed_data_3 = transformer_col_3.fit_transform(CGPA_COL)
sns.kdeplot(transformed_data_3,x='CGPA')
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
transformer_col_power = PowerTransformer(method='box-cox')
```

```python
transformed_data_power = transformer_col_power.fit_transform(CGPA_COL)
```

```python
sns.kdeplot(transformed_data_power)
```
