```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('../students.csv')
```

```python
sns.boxplot(df,x='IQ')
```

```python
sns.kdeplot(df,x='IQ')
```

```python
upper_outlier = df['IQ'].mean() + 3 + (df['IQ'].std())
lower_outlier = df['IQ'].mean() + 3 - (df['IQ'].std())
```

```python
df['IQ'] = np.where(df['IQ'] > upper_outlier,upper_outlier,np.where(df['IQ'] < lower_outlier,lower_outlier,df['IQ']))
```

```python
df['IQ']
```

```python
#IQR Method
upper_data = df['IQ'].quantile(0.75)
lower_data = df['IQ'].quantile(0.25)
```

```python
upper_range = upper_data + 1.5*(upper_data-lower_data)
lower_range = lower_data - 1.5*(upper_data-lower_data)
```
